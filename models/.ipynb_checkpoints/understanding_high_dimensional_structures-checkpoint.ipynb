# DeepTone

This is a notebook for understanding which structures are formed by potential clusters in Deeptone's embedding space. First it is important to understand Deeptone's outputs. 

Assuming that "deeptone_identity.yaml" is the traning configuration file, the model output of blocks 1 and 2 results in a set of identiy vectors (the number of which depends on the training batch size) of 128 dimensions each.

In block 3 each vector is normalised via the lambda function 'lambda x: tf.math.l2_normalize(x, axis=-1)'

In block 4 each normalised vector is then passed through a moving window, which averages its contents. This reduces the subsequent number of identity vectors equal to the number of original number of vectors per batch, divided by the size of the moving window. This is achived via the function: 'lambda x: tf.reduce_mean(tf.signal.frame(x[:, 8:, :], 4, 4, pad_end=False, axis=1), axis=2)' 

The loss of the resulting vectors are then processed using a semi-hard triplet loss function, using L2 as a distance metric.

The questions I attempt to answer here are:

What happens to the structure of clusters when we noramlise their vectors?

What happens to the normalised clusters when we average over them using a moving window?

I will only work with two dimensions, x and y, to simplify the visualisation of higher dimensions. x & y will be simulated using a multivariate gaussian. We will see how the final structures Deeptone is training on depend on the mean, variance and covariance of x and y.

```python
import numpy as np
import sklearn
import seaborn as sns
from matplotlib import pyplot as plt
import pandas as pd
import tensorflow as tf
from deeptone_classifier import *
import operator as op
from itertools import chain

def create_matrix(mean,var_fact,covv,n):
    """ Simplification of the 128 dimensional identity vectors that are used to calculate the loss in each training batch
    Create random mutlivariate gaussian with 2 dimensions: x & y
    Apply the exact same trasnformations found in blocks 3 and 4 of deeptone_identity.yaml
        

    Args:
        mean (list): List of mean parameters for centering x and y respectively
        var_fact (list): List of variances of x and y, expressed as a factor of their mean
        covvt (float): Covariance of x and y
        n (int): Number of pseudo-identity vectors to generate

    Returns:
        list: Dataframe representing the: (a)  pseudo-indentity vectors before normalisation, 
                                          (b) pseudo-indentity vectors afternormalisation, 
                                          (c) normalised pseudo-indentiy vectors after moving window average
    """    
    # Constat in case mean = 0
    const = 1
    
    # Create random mutlivariate gaussian with 2 dimensions: x & y. 
    # This is meant to be a simplification / simulation of the 128 dimensional identity vectors
    var_0 = abs(mean[0])*var_fact[0]+const
    var_1 = abs(mean[1])*var_fact[1]+const
    cov = [[var_0,covv],[covv,var_1]]
    
    # Here I create n "identity" embedding with only 2 dimensions.
    random = np.random.multivariate_normal(mean, cov, n)    
    random = random.reshape(1,random.shape[0],random.shape[1])
    
    # Calculation that is exactly equivalent to block 3 of deeptone_identity.yaml (see intro at beginning of notebook)
    random_l2 =  tf.math.l2_normalize(random, axis=-1)
    
    # Calculation that is exactly equivalent to block 4 of deeptone_identity.yaml (see intro at beginning of notebook)
    random_l2_mean = tf.reduce_mean(tf.signal.frame(random_l2, 4, 4, pad_end=False, axis=1), axis=2)

    # Re-formatting
    random_df = pd.DataFrame(data = np.array(random[0,:,:]), columns = ["x","y"])
    random_l2_df = pd.DataFrame(data = np.array(random_l2[0,:,:]), columns = ["x","y"])
    random_l2_mean_df = pd.DataFrame(data = np.array(random_l2_mean[0,:,:]), columns = ["x","y"])
    
    return random_df, random_l2_df, random_l2_mean_df


def plotting(random_df,random_l2_df,random_l2_mean_df,name):
    plt.close()
    f, axes = plt.subplots(2, 3,figsize=(15,15))
    sns.scatterplot(  y="y", x= "x", data=random_df,   ax=axes[0,0])
    axes[0,0].set_title("Before normalisation")

    sns.scatterplot(  y="y", x= "x", data=random_l2_df, ax=axes[0,1])
    axes[0,1].set_title("After normalisation")

    # Create unit circle
    theta = np.linspace(0, 2*np.pi, 100)
    r = np.sqrt(1.0)
    x1 = r*np.cos(theta)
    x2 = r*np.sin(theta)
    axes[0,2].plot(x1, x2,color="k",linewidth=0.5)


    sns.scatterplot(  y="y", x= "x", data=random_l2_mean_df, ax=axes[0,2])
    axes[0,2].set_title("After normalisation + \n moving average")

    sns.histplot( x= "value", hue="variable", data=pd.melt(random_df,value_vars=["x","y"]),   ax=axes[1,0])
    sns.histplot( x= "value", hue="variable", data=pd.melt(random_l2_df,value_vars=["x","y"]), ax=axes[1,1])
    sns.histplot( x= "value", hue="variable", data=pd.melt(random_l2_mean_df,value_vars=["x","y"]), ax=axes[1,2])
    axes[0,1].axis(xmin=-1.5,xmax=1.5,ymin=-1.5,ymax=1.5)
    axes[0,2].axis(xmin=-1.5,xmax=1.5,ymin=-1.5,ymax=1.5)
    plt.savefig(f'{name}.png')
```

# What happens to data clusters when we noramlise their vectors?

```python
# Generate: (a) random_df: pseudo-indentity vectors before normalisation, 
#           (b) random_l2_df: pseudo-indentity vectors afternormalisation, 
#           (c) random_l2_mean_df: normalised pseudo-indentiy vectors after moving window average
random_df, random_l2_df, random_l2_mean_df = create_matrix(mean=[0,0],var_fact=[0,0],covv=0,n=1000)

# Plot
plotting(random_df,random_l2_df,random_l2_mean_df,"1")

# How to read the plots: Each column represents each stage of processing of the identiy embeddings
# before they are used to calculate the loss

# In this plot we have:
# 1) [x , y] centred at [0,0]
# 2) Variance: intermediate 
# 3) Covariance: none

# After normalisation, the identity vectors are projected onto a unit circle. 
# The chances of lying anywhere on the circle are symmetric, but not equal. 
# Likeliest values are at x,y = (+-1,0) or (0,+-1)
# After averaging over noramlised vectors, unit circle is lost, all values now lie WITHIN it.
# The likeliest values are at the center of the circle
```

```python
# Next we simply shift the means of y.

# Generate: (a) random_df: pseudo-indentity vectors before normalisation, 
#           (b) random_l2_df: pseudo-indentity vectors afternormalisation, 
#           (c) random_l2_mean_df: normalised pseudo-indentiy vectors after moving window average
random_df, random_l2_df, random_l2_mean_df = create_matrix(mean=[0,1],var_fact=[0,0],covv=0,n=1000)


plotting(random_df,random_l2_df,random_l2_mean_df,"2")


# How to read the plots: Each column represents each stage of processing of the identiy embeddings
# before they are used to calculate the loss

# In this plot we have:
# 1) [x , y] centred at [0,1]
# 2) Variance: little  
# 3) Covariance: 0

# No change to shape of the multi-variate gaussian before normalisation, only a shift in mean.
# After normalisation identiy vectors are still projected onto a unit circle. 
# The chances of lying anywhere on the circle are no longer symmetric: Likeliest values are a the top of the circle.
# After averaging over noramlised vectors, unit circle is lost again, all values still lie WITHIN.
# The likeliest values are no longer at the center of the circle, they start to cluster at the top.

# Lets try by only shifting the mean of x
```

```python
# Next we simply shift the means of both x and y pseudo-identity vectors

# Generate: (a) random_df: pseudo-indentity vectors before normalisation, 
#           (b) random_l2_df: pseudo-indentity vectors afternormalisation, 
#           (c) random_l2_mean_df: normalised pseudo-indentiy vectors after moving window average
random_df, random_l2_df, random_l2_mean_df = create_matrix(mean=[1,0],var_fact=[0,0],covv=0,n=1000)


plotting(random_df,random_l2_df,random_l2_mean_df,"3")


# How to read the plots: Each column represents each stage of processing of the identiy embeddings
# before they are used to calculate the loss

# In this plot we have:
# 1) [x , y] centred at [1,0]
# 2) Variance: little to intermediate 
# 3) Covariance: 0

# The story:
# The likeliest values are no longer at the center of the circle, they start to cluster to the right

# Now lets use more extreme values
```

```python
# Next we simply shift the means of both x and y pseudo-identity vectors

# Generate: (a) random_df: pseudo-indentity vectors before normalisation, 
#           (b) random_l2_df: pseudo-indentity vectors afternormalisation, 
#           (c) random_l2_mean_df: normalised pseudo-indentiy vectors after moving window average
random_df, random_l2_df, random_l2_mean_df = create_matrix(mean=[10,0],var_fact=[0,0],covv=0,n=1000)


plotting(random_df,random_l2_df,random_l2_mean_df,"4")


# How to read the plots: Each column represents each stage of processing of the identiy embeddings
# before they are used to calculate the loss

# In this plot we have:
# 1) x and y centred at [10,0]
# 2) Variance: intermediate 
# 3) Covariance: 0


# The clustering becomes much more pronounced because:
# The mean of the distributon of each dimension dictates the position on the unit circle in which clustering occurs:
# The ratio of mean to variance dictates how well the cluster is (try increasing this value)
# As the variance increases, the more likely it is for averaged to enter the centre of the circle

# Now lets change the covariance
```

```python
# Next we simply shift the means of both x and y pseudo-identity vectors

# Generate: (a) random_df: pseudo-indentity vectors before normalisation, 
#           (b) random_l2_df: pseudo-indentity vectors afternormalisation, 
#           (c) random_l2_mean_df: normalised pseudo-indentiy vectors after moving window average
random_df, random_l2_df, random_l2_mean_df = create_matrix(mean=[10,0],var_fact=[0,0],covv=10,n=1000)


plotting(random_df,random_l2_df,random_l2_mean_df,"5")


# How to read the plots: Each column represents each stage of processing of the identiy embeddings
# before they are used to calculate the loss

# In this plot we have:
# 1) x and y centred at [10,0]
# 2) Variance: intermediate 
# 3) Covariance: 10

# The story:
# The covariance is a bit like the variance, it increases the size of the cluster, but mostly around the edge of the circel.
# Increasing the covariance play a lesser role than the variance, in increasing the chances of points lying closer to the centre
```

How can we extend this to higher dimensions. I would naively assume that the same behaviour holds for 3 or more dimensions: I.e that the direction and spread of clustering respectively is controled by the mean the values occuring in each dimensions, and a combination of the variance of that dimension with its covariance with other dimensions.

The best amount of clustering occurs when each cluster only appear in certain tight edges of the unit sphere. For this to occur, large means are required for the underlying distributions of each dimensions, and, litte variance and covariance. As the mean values decrease, and the (co)variance increase, the clusters becomes less pronounced. 

Interestingly, it seems that even if only one of the dimension is well pronounced, i.e. large enough mean, whilst having little variance and co-variance, this is enough to create well defined clusters, even if the other dimension is not well pronounced (i.e. small average, large variance). What is important is that the covariance remains small. As soon the covariance increases, two scenarios develop:

a) If both dimensions are not that well defined to beging with, the clustering simply becomes more noisy.

b) If one of the dimension is well defined, but the other is not, increasing the covariance creates many mini clusters which may be confused by independent clusters if an algorithm were to sensitive: 

```python
# Generate: (a) random_df: pseudo-indentity vectors before normalisation, 
#           (b) random_l2_df: pseudo-indentity vectors afternormalisation, 
#           (c) random_l2_mean_df: normalised pseudo-indentiy vectors after moving window average
random_df, random_l2_df, random_l2_mean_df = create_matrix(mean=[10,10],var_fact=[1000,1],covv=100,n=1000)

# Plot
plotting(random_df,random_l2_df,random_l2_mean_df,"6")
```

It is hard to imagine what other strange shapes may occur as we increase the dimensions, and possible how sub-clusters may appear due to the interaction of dimensions. Indeed this scenario is problematic because although the clusters may be well defined before normalisation, post normalisation and averaging may completely disrupt this clustering.

There, to quantify clustering one must consider the depth, angle and spread of each cluster, such that they avoid as much interesction as possible. What we could do, therefore, is measure the angle of the centre of each cluster, by measuring the angle of the direction of each dimension relative to another fixed dimension. The coordinates of the center will then give us a sense of depth, and the quartiles of each dimension give us the spread.


Important notes: When applying the moving average on REAL data, there are bound to be some temporal correlations that are being ommited in this data set. It would probabaly be make more sense to use a Gaussian process to simulate each dimension.

The most informative measures would be the degree of correlation of the identiy matrices BEFORE they are normalised.
